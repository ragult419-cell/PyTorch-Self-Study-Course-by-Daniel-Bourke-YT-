```python
import numpy as np
import matplotlib.pyplot as plt

# Create the weight vector (fixed)
w = np.array([2, 1])

# Try multiple input vectors (change the angle manually)
angles = np.linspace(0, 2 * np.pi, 16)
origin = np.array([[0, 0], [0, 0]])  # for quiver plot origin

plt.figure(figsize=(10, 10))
plt.quiver(*origin, w[0], w[1], color='r', scale=1, scale_units='xy', angles='xy', label='w (weight vector)')

# Loop through different x vectors at various angles
for angle in angles:
    x = np.array([np.cos(angle), np.sin(angle)])
    dot_product = np.dot(w, x)
    color = 'g' if dot_product > 0 else ('b' if dot_product < 0 else 'orange')
    plt.quiver(*origin, x[0], x[1], color=color, scale=1, scale_units='xy', angles='xy')
    plt.text(x[0]*1.2, x[1]*1.2, f"{dot_product:.2f}", fontsize=8)

plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.grid(True)
plt.axis('equal')
plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.title("Dot Product Sign with Varying Vector Angles\n(Green = positive, Blue = negative, Gray = zero)")
plt.legend()
plt.show()
```

```python
y = np.hstack((-1 * np.ones(50), +1 * np.ones(50)))
```

```python
w = np.random.randn(2)
```

```python
b = 0
```

```python
lr = 0.1  # learning rate
```

```python
for epoch in range(10):  # Loop over the dataset 10 times
    for i in range(len(X)):  # Loop over each data point
        xi = X[i]
        yi = y[i]
        prediction = np.sign(np.dot(w, xi) + b)
        if prediction != yi:
            w += lr * yi * xi
            b += lr * yi
```
