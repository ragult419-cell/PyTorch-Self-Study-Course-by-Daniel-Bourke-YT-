## Reshaping, stacking, squeezing, and unsqueezing

- Reshaping: reshapes an input tensor to a defined shape
- View: View of an input tensor of certain shape but keep the same memory as the original tensor
- Stacking: Combine multiple tensors on top of each other (vstack) or side by side (hstack)
- Squeeze: removes all 1 dimension from a tensor
- Unsqueeze: Add a 1 dimension to a target tensor
- Permute: Return a view of the input with dimensions permuted (swapped) in a certain way.


```python
# Let's create a tensor
import torch
x = torch.arange(1., 10.)
x, x.shape
```

```python
# Add an extra dimension
x_reshaped = x.reshape(1, 9)
x_reshaped, x_reshaped.shape
```

```python
# Change the view
z = x.view(1, 9)
z, z.shape
```

```python
# Changing z changes x (because a view of a tensor shares the same memory as the original)
z[:, 0] = 5
z, x
```

```python
# Stack tensors on top of each other
x_stacked = torch.stack([x, x, x, x], dim=0)
x_stacked
```

```python
# torch.squeeze() - removes all single dimensions from a target tensor
print(f"Previous tensor: {x_reshaped}")
print(f"Previous shape: {x_reshaped.shape}")


#Remove extra dimensions from x_reshaped
x_squeezed = x_reshaped.squeeze()
print(f"\nNew tensor: {x_squeezed}")
print(f"New shape: {x_squeezed.shape}")
```

```python
# torch.unsqueeze() - add a single dimension to a target tensor at a specific dim
print(f"Previous tensor: {x_squeezed}")
print(f"Previous shape: {x_squeezed.shape}")

x_unsqueezed = x_squeezed.unsqueeze(dim=0)
print(f"\nNew tensor: {x_unsqueezed}")
print(f"New shape: {x_unsqueezed.shape}")
```

```python
# torch.permute - rearranges the dimensions of a target tensor in a specific order
x_original = torch.rand(size = (224, 224, 3)) # [height, width, colour_channels]

#Permute the original tensor to rearrange the axis (or dim) order
x_permuted = x_original.permute(2, 0, 1) #shifts axis 0 to 1, 1 to 2, 2 to 0

print(f"Previous Shape: {x_original.shape}")
print(f"New Shape: {x_permuted.shape}") # [colour_channels, height, width]
```

## Indexing (selecting data from tensors)

Indexing with PyTorch is similar to indexing with Numpy

```python
# Create tensor
y = torch.arange(1, 10).reshape(1, 3, 3)
y, y.shape
```

```python
# Let's index on our new tensor
y[0]
```

```python
# Lets index on the middle bracket (dim=1)
y[0][0]
```

```python
# Lets index on the most inner bracket (last dimension)
y[0][0][0]
```

```python
# Getting 9, practice
y[0][2][2]
```

```python
# You can also use ":" to select "all" of a target dimension
y[:, 0]
```

```python
# Get all values of 0th and 1st dimensions but only index 1 of the 2nd dimension
y[:, :, 1]
```

```python
# Get all values of the 0 dimension but only the 1 index value of 1st and 2nd dimension
y[:, 1, 1]
```

```python
# Get index 0 of 0th and 1st dimension and all values of 2n dimension
y[0, 0, :]
```

```python
# Index on y to return 9
y[0, 2, 2]
```

```python
# Index on y to return 3, 6, 9

y[:, :, 2]
```

## PyTorch tensors and NumPy

NumPy is a popular scientific Python numerical computing library
And because of this, PyTorch has functionality to interact with it.
* Data in NumPy, want in PyTorch tensor - torch.from_numpy(ndarray)

*Data in torch to numPy - torch.Tensor.numpy()

```python
# NumPy array to tensor
# Default dtype for NP is float64 while a torch tensor default is float32
# Warning - when converting from a NP to a torch tensor, the default will be float64 unless changed
import numpy as np

array = np.arange(1.0, 8.0)
c = torch.from_numpy(array)
array, c
```

When converting from np to tensor or other way around, changes to original array or tensor does not change the new tensor or array

Any new tensor or array will reflect the dtype of the original array or tensor

### To reduce the randomness in neural networks and PyTorch comes the concept of a **random seed**

Essentially what the random seed does is "flavour" the randomness.

```python
## Reproducbility (trying to take random out of random)

# Create two random tensors
rand_tensor_A = torch.rand(3,4)
rand_tensor_B = torch.rand(3,4)

print(rand_tensor_A)
print(rand_tensor_B)
print(rand_tensor_A == rand_tensor_B)
```

```python
# Let's make some random but reproducable tensors

#Set random seed
RANDOM_SEED = 42

torch.manual_seed(RANDOM_SEED)
random_tensor_C = torch.rand(3, 4)

torch.manual_seed(RANDOM_SEED)
random_tensor_D = torch.rand(3, 4)

print(random_tensor_C)
print(random_tensor_D)
print(random_tensor_C == random_tensor_D)
```
